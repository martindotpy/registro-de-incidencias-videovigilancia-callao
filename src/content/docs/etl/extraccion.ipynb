## Extracción


Se realiza la extracción de datos desde la fuente original, en este caso un
archivo CSV que será descargado desde la plataforma de datos abiertos del Estado
Peruano, ubicado en la siguiente URL:
<https://www.datosabiertos.gob.pe/dataset/registro-de-incidencias-observadas-mediante-la-central-de-monitoreo-de-c%C3%A1maras-de-11>.


```python
import curl_cffi
from IPython.display import Markdown

from registro_de_indicencias.folder import (
    CSV_PATH,
    DATA_FOLDER_PATH,
    WORKSPACE_FOLDER_PATH,
)
```

```python
DATA_FOLDER_PATH.mkdir(exist_ok=True)

if not CSV_PATH.exists():
    csv_response = curl_cffi.get(
        "https://www.datosabiertos.gob.pe/sites/default/files/REGISTRO%20DE%20INCIDENCIAS%20OBSERVADAS%20MEDIANTE%20LA%20CENTRAL%20DE%20MONITOREO%20DE%20C%C3%81MARAS%20Y%20VIDEOVIGILANCIA%20CALLAO-MPC-v1.3.csv",
        stream=True,
    )

    with CSV_PATH.open("wb") as f:
        for chunk in csv_response.iter_content():
            f.write(chunk)
```

```python
display(
    Markdown(
        f"""\
El archivo descargado se ha guardado en la ruta: `{CSV_PATH.relative_to(WORKSPACE_FOLDER_PATH)}`. A continuación,
se muestra información relevante sobre el archivo:

- **Nombre del archivo:** `{CSV_PATH.name}`
- **Tamaño del archivo:** ${CSV_PATH.stat().st_size / (1024 * 1024):.2f} \\text{{MB}}$"""
    )
)
```
